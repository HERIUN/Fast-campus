## package

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'source/dog.jpg'

image_pil=Image.open(path)
image = np.array(image_pil)
```

## 이미지 들여다보기
 - 이미지 열기전 shape 및, min,max를 통해 이미지의 range확인

```python
image.shape
```

```python
np.max(image), np.min(image) 
```

## 그래프로 시각화

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

## 이미지 흑백으로 열기

```python
image_bw = image_pil.convert('L')
image_bw = np.array(image_bw)
image_bw.shape
```

```python
plt.imshow(image_bw)
plt.show()
```

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

```python
plt.imshow(image_bw, 'RdBu')
plt.show()
```

```python
plt.imshow(image_bw, 'jet')
plt.show()
```

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10, 20))
plt.imshow(image)
plt.show()
```
